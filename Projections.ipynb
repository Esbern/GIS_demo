```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# prompt: load a geojson data set from my public github repo https://github.com/Esbern/gis_demo_data/blob/main/europa.geojson


# Load the GeoJSON data from the GitHub repository
df_europa = gpd.read_file('https://github.com/Esbern/gis_demo_data/blob/main/europa.geojson?raw=true')

# Plot the data
df_europa.plot()
plt.show()
```

```python
df = df_europa.to_crs(epsg=25832)
df.plot()
```

```python
# prompt: zoon in on denmark uising df_europa and do  not remove naighbering countries

import matplotlib.pyplot as plt
# Filter the GeoDataFrame to include only Denmark and its neighboring countries
denmark_and_neighbors = df_europa[df_europa['NAME'].isin(['Denmark', 'Germany', 'Sweden', 'Norway', 'Poland'])]

# Calculate the bounds of Denmark
denmark_bounds = df_europa[df_europa['NAME'] == 'Denmark'].bounds

# Set the plot limits to zoom in on Denmark and its neighbors
minx, miny, maxx, maxy = denmark_bounds.minx.iloc[0], denmark_bounds.miny.iloc[0], denmark_bounds.maxx.iloc[0], denmark_bounds.maxy.iloc[0]

# Plot the zoomed-in map
denmark_and_neighbors.plot()
plt.xlim(minx, maxx)
plt.ylim(miny, maxy)
plt.show()
```
